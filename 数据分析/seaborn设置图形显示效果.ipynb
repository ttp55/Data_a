```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pandas import Series, DataFrame
```

```python
x = np.linspace(0,14,100)
```

```python
y1 = np.sin(x)
y2 = np.sin(x+2)*1.25
```

```python
def sinplot():
    plt.plot(x,y1)
    plt.plot(x,y2)
```

```python
sinplot()
```

```python
sns.set_style('darkgrid', {'grid.color':'black'})
```
