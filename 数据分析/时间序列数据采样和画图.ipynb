```python
t_range = pd.date_range('2016-01-01','2016-12-31', freq='D')
```

```python
t_range
```

```python
stock_df = DataFrame(index=t_range)
```

```python
stock_df['AL'] = np.random.randint(80, 160, size=len(t_range))
```

```python
stock_df['TX'] = np.random.randint(30, 60, size=len(t_range))
```

```python
stock_df.head()
```

```python
stock_df.plot()
```

```python
weekly_df = DataFrame()
```

```python
weekly_df['AL'] = stock_df['AL'].resample('W').mean()
```

```python
weekly_df['TX'] = stock_df['TX'].resample('W').mean()
```

```python
weekly_df.plot()
```
